**Aggregation**

You‚Äôre predicting the severity of an accident ‚Äî one label per accident.
So in the end, you want one row = one accident in your training data.

BUT‚Ä¶
our data isn‚Äôt shaped that way üòÖ


why we use aggregation 

Because users and vehicles files have multiple rows per accident,
if you merge them directly ‚Äî you‚Äôll get duplicate accident rows 

If you train like this, the model will think there are two accidents ‚Äî but actually it‚Äôs one!
That‚Äôs wrong.

**‚öôÔ∏è Solution: Aggregate before merging**

Aggregation = compressing many rows into one summary per accident
You summarize numeric and categorical values into features that represent the whole accident.

**Many Vehicles per accidents --- needs aggregation** 

here's where aggregation matter the most 

- Accident Id important because we can add all together because of accident id 

- Direction - movement direction 

- speed_mean - mean speed - Avg speed 

- speed_max - max(speed) - More speed = danger 

- Category - type of vehicles 

- PassengerNumber - how many passengers

- ImpactPoint 

many more but i got now we can aggregate only if we have that data in our dataset like speed mean , max not in my dataset but it's good if we have anyways let work with what we ahve 

```python
vehicles_agg = vehicles.groupby('AccidentId').agg(
    num_vehicles=('VehicleId', 'nunique'),
    passenger_mean=('PassengerNumber', 'mean'),
    vehicle_types=('Category', 'nunique'),
    direction_mode=('Direction', lambda x: x.mode().iloc[0] if not x.mode().empty else "Unknown"),
    mobile_obstacle_count=('MobileObstacle', 'nunique'),
    fixed_obstacle_count=('FixedObstacle', 'nunique'),
    impact_point_mode=('ImpactPoint', lambda x: x.mode().iloc[0] if not x.mode().empty else "Unknown"),
    maneuver_mode=('Maneuver', lambda x: x.mode().iloc[0] if not x.mode().empty else "Unknown")
).reset_index()
```

```python
main['AccidentId'] = main['AccidentId'].astype(str)
vehicles_agg['AccidentId'] = vehicles_agg['AccidentId'].astype(str)

main = main.merge(vehicles_agg, on='AccidentId', how='left')
```

```python
# --- Clean SafetyDeviceUsed ---
users['SafetyDeviceUsed'] = (
    pd.to_numeric(users['SafetyDeviceUsed'], errors='coerce')  # convert 'yes', 'no', etc. to NaN
    .fillna(0)  # assume missing = not used (can also skip if you prefer NaN)
)

# --- Clean BirthYear ---
users['BirthYear'] = pd.to_numeric(users['BirthYear'], errors='coerce')

# --- Optional cleanup for gender ---
users['Gender'] = users['Gender'].replace({
    'Homme': 'M', 'Male': 'M', 'Femme': 'F', 'Female': 'F', 1: 'M', 0: 'F'
})

# --- Aggregate safely ---
user_agg = users.groupby('AccidentId').agg(
    num_vehicles=('VehicleId', 'nunique'),
    seats_list=('Seat', lambda x: ', '.join(x.dropna().unique()) if len(x.dropna()) > 0 else 'Unknown'),
    common_category=('Category', lambda x: x.mode().iloc[0] if not x.mode().empty else "Unknown"),
    common_gender=('Gender', lambda x: x.mode().iloc[0] if not x.mode().empty else "Unknown"),
    common_tripreason=('TripReason', lambda x: x.mode().iloc[0] if not x.mode().empty else "Unknown"),
    num_safetydevice=('SafetyDevice', 'nunique'),
    safety_used_ratio=('SafetyDeviceUsed', 'mean'),
    avg_birthyear=('BirthYear', 'mean')
).reset_index()
```

```python
main['AccidentId'] = main['AccidentId'].astype(str)
user_agg['AccidentId'] = user_agg['AccidentId'].astype(str)
main = main.merge(user_agg, on='AccidentId', how='left')
```

#### filling the Data 

We remove too much missing data columns but we have some columns with little bit missing values so just filling with median or mode values 

```python
from sklearn.impute import SimpleImputer

# separate numeric and categorical columns 

num_cols = main.select_dtypes(include=['int64','float64']).columns
cat_cols = main.select_dtypes(include=['object']).columns

# Imputer 

num_imputer = SimpleImputer(strategy='median')
cat_imputer = SimpleImputer(strategy='most_frequent')


# Apply and transform 

main[num_cols] = num_imputer.fit_transform(main[num_cols])
main[cat_cols] = cat_imputer.fit_transform(main[cat_cols])
```
